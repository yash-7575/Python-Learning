```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4]
y=[2,4,6,8]
plt.plot(x,y) #plot is used for line chart
plt.title("Simple Line Graph")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
```

```python
import numpy as np

x = np.linspace(0, 10, 20)
y = np.sin(x)

plt.scatter(x, y, color='red', marker='o')
plt.title("Scatter Plot of Sine Function")
plt.show()
```

```python
x=np.linspace(10,100,10) 
'''linspace = linear space '''
''' np.linespace(start,stop, num) evenly distribute the array into num parts'''
''' in this example it will generate an array from 10 to 100 and it will divide that array into 10 equal parts'''
x
```

```python
x=np.linspace(10,100,10, endpoint=False)
x
```

```python
x=np.linspace(0,10,20)
y=np.sin(x)
plt.scatter(x,y,color="red",marker="x")
plt.title("Sine fn")
```

```python
plt.subplot(1, 2, 1)  # 1 row, 2 columns, 1st plot
plt.plot([1, 2, 3], [4, 5, 6], 'r')  # Red line

plt.subplot(1, 2, 2)  # 1 row, 2 columns, 2nd plot
plt.bar([1, 2, 3], [4, 5, 6], color='blue')

plt.show()
```

```python
categories = ["A", "B", "C", "D"]
values = [3, 7, 1, 8]

plt.bar(categories, values, color=['red', 'blue', 'green', 'orange'])
plt.title("Bar Chart Example")
plt.show()
```

```python
x=['A','B','C','D']
y=[10,5,15,7]
plt.bar(x,y)
```

```python
x=['A','B','C','D']
y=[10,5,15,7]
plt.bar(x,y,color=['red','green','blue','yellow'])
plt.title("Bar Chart")
plt.xlabel("Company")
plt.ylabel("Stock Price")
```

```python
data = np.random.randn(1000)  # 1000 random numbers

plt.hist(data, bins=30, color='purple', edgecolor='black')
plt.title("Histogram Example")
plt.show()
```

```python
data = np.arange(11)
plt.hist(data, bins=30)  # Specify number of bins
'''Histograms are simply bar charts but the special feature is
it varies with the frequency of data, suppose 20 is repeated 5 times then y value of graph for 20 is 5'''
```

```python
data=[1,1,2,2,2,2,3,3,3,4,4,4,4,4]
plt.hist(data,bins=10) #groups the data into ranges
```
